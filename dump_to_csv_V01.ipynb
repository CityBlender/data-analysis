```python
folders = ['americas',
           'asia',
           'eur',
           'oceania',
           'uk']
root = os.getcwd()
```

```python
def create_FuinkiPanda(folder = folders[1]):
    Funki_Panda = []
    print(folder)
    print('******************')
    
    # Fetch the selected folder
    # And target that dir
    sub_fldr = (os.getcwd() + '\\' + folder)  
    os.chdir(sub_fldr)
    print(sub_fldr)
    
    # Create a list of all the .json files inside the selected folder
    json_list = glob.glob("*.json")
    
    # Then loop inside each file 
    for j in json_list:        
        test_json_file = sub_fldr + '\\' + j
        print(test_json_file)
        
        json_data = json.load(open(test_json_file))
        
        if json_data['resultsPage']['totalEntries'] == 0:
            pass
        
        else:
            df = pd.DataFrame(json_data['resultsPage']['results']['event'])
        
            # Append the dataset
            Funki_Panda.append(df)
    
    
    
    Funki_Panda = pd.concat(Funki_Panda, axis=1)
    result_fldr = (root + '\\CSVs')
    os.chdir(result_fldr)
    print(os.getcwd())
    
    Funki_Panda.to_csv(folder + '.csv')
    
    
    # Go back to original folder
    os.chdir(root)
    return(Funki_Panda)


```

```python
for i in range(len(folders)):
    print(i)
    create_FuinkiPanda(folder = folders[i])
    
```
